```python
class Human: #クラス定義
    def __init__(self, name, age):
        self.name = name 
        self.age = age   

    def check_adult(self):    #メゾット部分をif文に変更
        if self.age >= 20:    #年齢制約
            print(f"{self.name}は大人です")
        else:
            print(f"{self.name}は大人ではありません")

#リスト作成
human_list = [
    Human("太郎", 25),
    Human("金太郎", 18),
    Human("花子", 30)
]

#リストの中身をチェック
for human in human_list:
    human.check_adult()
```
